```python
df_vehicle.reset_index(drop=True, inplace=True)
```

```python
valores_unicos = df_vehicle['STATE_REGISTRATION'].unique()

# Crear un diccionario de mapeo de valores únicos a valores numéricos
mapeo_SR = {valor: indice for indice, valor in enumerate(valores_unicos)}

# Aplicar el mapeo a la columna del DataFrame
df_vehicle['STATE_REGISTRATION'] = df_vehicle['STATE_REGISTRATION'].map(mapeo_SR )
```

```python
valores_unicos = df_vehicle['DRIVER_LICENSE_STATUS'].unique()

# Crear un diccionario de mapeo de valores únicos a valores numéricos
mapeo_LS = {valor: indice for indice, valor in enumerate(valores_unicos)}

# Aplicar el mapeo a la columna del DataFrame
df_vehicle['DRIVER_LICENSE_STATUS'] = df_vehicle['DRIVER_LICENSE_STATUS'].map(mapeo_LS)
```

```python
valores_unicos = df_vehicle['POINT_OF_IMPACT'].unique()

# Crear un diccionario de mapeo de valores únicos a valores numéricos
mapeo_PI = {valor: indice for indice, valor in enumerate(valores_unicos)}

# Aplicar el mapeo a la columna del DataFrame
df_vehicle['POINT_OF_IMPACT'] = df_vehicle['POINT_OF_IMPACT'].map(mapeo_PI)
```

```python
valores_unicos = df_vehicle['DRIVER_LICENSE_JURISDICTION'].unique()

# Crear un diccionario de mapeo de valores únicos a valores numéricos
mapeo_LJ = {valor: indice for indice, valor in enumerate(valores_unicos)}

# Aplicar el mapeo a la columna del DataFrame
df_vehicle['DRIVER_LICENSE_JURISDICTION'] = df_vehicle['DRIVER_LICENSE_JURISDICTION'].map(mapeo_LJ)
```

```python
valores_unicos = df_vehicle['VEHICLE_TYPE'].unique()

# Crear un diccionario de mapeo de valores únicos a valores numéricos
mapeo_VT = {valor: indice for indice, valor in enumerate(valores_unicos)}

# Aplicar el mapeo a la columna del DataFrame
df_vehicle['VEHICLE_TYPE'] = df_vehicle['VEHICLE_TYPE'].map(mapeo_VT)
```

```python
#MLPRegressor
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
X, y = make_classification(n_samples=100, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    random_state=1)
clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
clf.predict_proba(X_test[:1])
clf.predict(X_test[:5, :])
clf.score(X_test, y_test)
```
